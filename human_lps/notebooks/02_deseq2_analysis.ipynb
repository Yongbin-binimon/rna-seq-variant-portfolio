```python
#DESeq2 분석은 정상적으로 실행됨
#↓
#유의 DEG는 XIST 1개
#↓
#대표 LPS 면역 유전자 IL6, CXCL8, NFKBIA는 유의하지 않음
#↓
#TNF 후보 ID들을 추가 확인해야 함
#↓
#샘플 조건/metadata 검증이 필요함
```

```python
import pandas as pd

# 1. Load DESeq2 result files
deg_results = pd.read_csv("../results/tables/deseq2_results_LPS_vs_Control.csv")
sig_deg = pd.read_csv("../results/tables/significant_DEG_padj_0.05.csv")

# 2. Clean column name
deg_results = deg_results.rename(columns={"Unnamed: 0": "Geneid"})
sig_deg = sig_deg.rename(columns={"Unnamed: 0": "Geneid"})

# 3. Define immune-response marker genes
immune_gene_map = pd.DataFrame({
    "Geneid": [
        "ENSG00000136244",  # IL6
        "ENSG00000169429",  # CXCL8
        "ENSG00000100906",  # NFKBIA
        "ENSG00000232810"   # TNF, main candidate from biomaRt result
    ],
    "gene_symbol": [
        "IL6",
        "CXCL8",
        "NFKBIA",
        "TNF"
    ]
})

# 4. Merge immune gene list with DESeq2 results
immune_deg_summary = immune_gene_map.merge(
    deg_results,
    on="Geneid",
    how="left"
)

# 5. Select key columns
immune_deg_summary = immune_deg_summary[
    [
        "gene_symbol",
        "Geneid",
        "baseMean",
        "log2FoldChange",
        "lfcSE",
        "stat",
        "pvalue",
        "padj"
    ]
]

# 6. Save summary table
immune_deg_summary.to_csv(
    "../results/tables/immune_gene_DESeq2_summary.csv",
    index=False
)

# 7. Display results
print("Number of total genes:", deg_results.shape[0])
print("Number of significant DEGs:", sig_deg.shape[0])

display(sig_deg)
display(immune_deg_summary)
```
