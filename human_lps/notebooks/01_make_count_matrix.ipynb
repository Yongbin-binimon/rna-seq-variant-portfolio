```python
import glob

count_files = sorted(glob.glob("../results/counts/*.counts.txt"))

print("count 파일 개수:", len(count_files))
for f in count_files:
    print(f)
```

```python
import pandas as pd

test_df = pd.read_csv(count_files[0], sep="\t", comment="#")
test_df.head()
```

```python
import os
import pandas as pd

dfs = []

for file in count_files:
    sample_name = os.path.basename(file).replace(".counts.txt", "")
    
    df = pd.read_csv(file, sep="\t", comment="#")
    df = df[["Geneid", df.columns[-1]]]
    df.columns = ["Geneid", sample_name]
    
    dfs.append(df)

count_matrix = dfs[0]

for df in dfs[1:]:
    count_matrix = count_matrix.merge(df, on="Geneid")

count_matrix.head()
```

```python
count_matrix.shape
```

```python
output_path = "../results/count_matrix.csv"

count_matrix.to_csv(output_path, index=False)

print("저장 완료:", output_path)
```
